### 7.1 확률적 데이터와 확률변수
 - 결정론적 데이터 : 언제 누가 얻더라도 항상 같은 값이 나오는 데이터
 - 확률적 데이터 : 혈압 처럼 정확이 예측할 수 없는 값이 나오는 데이터
    - 여러 조건이나 상황에 따라 데이터 값이 영향을 받을 수 있다.
    - 측정 시에 발생하는 오차 때문
    
   

- 분포

    - 범주형 데이터 > 카운트 플롯, 실수형 데이터 > 히스토그램

- 기술통계
    - 분포를 표현하는 또다른 방법
    - 숫자로서 분포를 나타낸다.
        - 표본평균, 표본중앙값, 표본최빈값
        - 표본분산, 표본표준편차
        - 표본왜도, 표본첨도

- 표본 평균
    - 확률적인 데이터값을 대표하는 기술통계 값 중 하나
    - 정확히는 '표본평균(sample mean, sample average)
    - 표본평균은 데이터 분포의 대략적인 위치를 나타낸다.
    - m, x^bar 를 사용

$$m = \bar x = \frac{1}{m}\sum_{i=1}^{N}x_i$$

이 식에서 N 은 자료의 개수, i는 자료의 순서(index)를 나타낸다.

- 표본중앙값
    - 전체 자료를 크기별로 정렬했을 때 가장 중앙에 위치하는 값.
    - 전체 표본 개수가 N인 경우, 
        - N이 홀수이면 중앙값은 (N+1)/2번째 표본의 값
        - N이 짝수이면 중앙값은 N/2 번째의 표본의 값과 N/2+1 번째 표본의 값의 평균

- 표본 최빈값(most frequent value, sample mod)
    - 데이터 값 중 가장 빈번하게 나오는 값
    - 유한한 종류의 값은 최빈값을 쉽게 구함
    - 혈압, 키, 몸무게 등 연속적인 값을 가지는 데이터(정밀도가 무한히 높다는 가정하)는 똑같은 값이 나올 확률이 아주 작기 때문에 최빈값을 구하기 어렵다. 
    - 연속적인 값의 경우에 일정한 구간 간격으로 나누어 가장 많은 데이터를 가진 구간의 대푯값을 그 데이터의 최반값으로 가정하는 방법을 많이 사용

- 파이썬을 사용한 대푯갑 계산
    - Numpy 함수를 사용하여 각종 대푯값 계산
    - mean, median, argmax, histogram

- 대칭 분포
    - 분포가 표본평균을 기준으로 대칭인 대칭 분포이면 표본중앙값은 표본평균과 같다.
    - 분포가 대칭분포이며 하나의 최고값을 가지는 단봉분포이면 표본 최빈값은 표본평균과 같다.
    - 대칭 분포를 비대칭으로 만드는 데이터가 더해지면 표본평균이 가장 크게 영향을 받고 표본 최빈값이 가장 적게 영향을 받는다.

```python
np.random.seed(1)
x = np.random.normal(size=1000)
x = np.hstack([x, 5 * np.ones(50)])
bins = np.linspace(-6, 6, 12 * 4  + 1)

sample_mean = np.mean(x)
sample_median = np.median(x)
mode_index = np.argmax(ns)
sample_mode = 0.5 * (bins[mode_index])

sns.distplot(x, bins=bins)
plt.axvline(sample_mean, c='k', ls = ':', label = "표본평균")
plt.axvline(sample_median, c='k', ls='--', label = "표본중앙값")
plt.axvline(sample_mode, c='k', ls="-", label = "표본최빈값")
plt.title("표본평균, 표본중앙값, 표본최빈값의 차이")
plt.xlabel("x")
plt.legend()
plt.show()
```

```python
x = df[df.species == "virginica"]["petal length (cm)"]

def sample_stastics(x, n, x_min=0, x_max= 10):
    bins = np.linspace(x_min, x_max, n + 1)
    ns, _ = np.histogram(x, bins=bins)
    
    sample_mean = np.mean(x)
    sampe_median = np.median(x)
    sampe_index = np.argmax(ns)
    sampe_mode = 0.5 * (bins[mode_index] + bins[mode_index + 1])
    print("구간이 {}개일 때".format(n))
    print("표본평균", sample_mean)
    print("표본중앙값", sampe_median)
    print("표본최빈값", sample_mode)
    
    sns.distplot(x, bins=bins)
    plt.axvline(sample_mean, c='k', ls=":", label="표본평균")
    plt.axvline(sample_median, c='k', ls="--", label=" 표본중앙값")
    plt.axvline(sample_mode, c='k', ls="-", label="표본최빈값")
    plt.title("표본평균, 표본중앙값, 표본최빈값의 차이")
    plt.xlabel("x")
    plt.legend()
    plt.show()
```

- 분산과 표준편차
    - 데이터가 얼마나 변동하는지 알려주는 특징값
    - 분포의 폭을 대표하는 값
    - 분산에 대한 양의 제곱근 값
    
    표준편차 = $\sqrt{분산}$
    
    표본분산 
    
    $$s^2 = \frac{1}{N}\sum_{i=1}^{N}(x_i - \bar{x})^2$$


- 자료값과 평균 사이의 거리를 의미
- 평균보다 작을 때는 음수가 나오므로 제곱을 하여 양수로 만듬
- 편향 표본분산 이라 한다.

- 비 편향 표본분산

$$s^2_ unbiased = \frac{1}{N-1}\sum_{i=1}^{N}(x_i - \bar{x})^2$$

- 파이썬을 사용한 표본 분산 및 표본표준편차의 계싼
    - var(), std(), ddof 인수를 이용하여 비편향 표본분산과 표본표준편차를 구할 수 있다.

표본비대칭도 = $$\frac{\tfrac{1}{N} \sum_{i=1}^N (x_i-\overline{x})^3}{ \sqrt{\tfrac{1}{N-1} \sum_{i=1}^N (x_i-\overline{x})^2}^{\,3}}$$

- 평균과의 거리가 세제곱을 이용하여 구한 특징값
- 0이면 분포가 대칭
- 음수면 표본평균값을 기준으로 왼쪽에 있는 값을 가진 표본이 나올 가능성이 더 많다.


- 표본첨도
    - 평균과의 거리의 네제곱을 이용하여 구한 특징값
    - 데이터가 중앙에 몰려있는 정도를 정밀하게 비교

```python
sp.stats.skew(x), sp.stats.kurtosis(x)
```

- 표본 모멘트
    - 분산, 비대칭도, 첨도를 구하기 위해 제곱, 세제곱, 네제곱을 하는 것처럼 k제곱을 이용하여 구한 모멘트를 k차 모멘트라 한다.
    
    - 표본 모멘트 = $\frac{1}{N}\sum_{i = 1}^{N} x^k_i$
    - 표본 중앙 모멘트 = $\frac{1}{N}\sum_{i = 1}^{N} (x_i - \bar{x})^k$
    - 평균은 1차 모멘트, 분산은 2차 모멘트, 비대칭 3차,첨도 4차
    - moment() 명령어

```python
sp.stats.moment(x, 1), sp.stats.moment(x, 2), sp.stats.moment(x, 3), sp.stats.moment(x, 4)
```

- 확률변수
    - 수학적으로 확률공간의 표본을 입력으로 받아서 실수인 숫자로 바꾸어 출력하는 함수'
    - 출력되는 실수가 데이터의 값
    - 표본이 실수여야 확률분포함수를 정의할 수 있다.

- 이산확률 변수
    - discrete 하도록 정의할 수 있는 변수
    - 표본 공간 원소 개수가 무한대인 경우도 이산 확률변수 가능

- 연속확률변수
    - 연속적이고 무한대의 실수 표본값을 가지는 확률변수
    - 모든 사건은 구간사건의 조합으로 표시된다.
    - 확률이 어떻게 할당되었는가를 나타내는 정보

- 확률변수는 데이터 생성기
    - 데이터를 생성하는 무언가가 존재한다고 가정할 수 있다.
    - 현실 세계의 데이터는 확률변수가 가진 확률분포에 따라 실수 표본공간에서 선택된 표본
    - 실현, 표본화 : 확률분포함수에 따라 표본공간의 표본이 현실세계의 데이터로 선택되는 것. 많은 수의 데이터 집합에서 일부 데이터만 선택하는 과정도 포함
    - 실현 : 이상적이고 추상적이며 수학적인 세계에서 현실로 넘어가는 과정. 
    - 확률변수에 따르면 현실의 표본데이터는 이러한 수학적인 세계가 현실 세계에 투영된 그림자에 지나지 않는다.
    
- 확률 변수의 특징
    - 확률변수로부터 데이터를 여러번 생성하는 경우 데이터 값은 변할 수 있지만 확률변수 자체는 변하지 않는다.
    - 확률분포함수는 직접 관찰할 수 없다. 실제 데이터 값을 이용하여 추정할 뿐
    - 확률변수에서 만들어지는 실제 데이터값은 확률변수가 가진 특성을 반영하고 있다. 데이터 개수가 증가할 수록 정확하게 확률분포 함수를 묘사할 수 있다.
    
- 확률변수를 사용한 데이터 분석
    - 데이터 수집 -> 수집 데이터를 어떤 확률변수의 표본 데이터라 가정 -> 해당 확률변수의 확률분포함수 모양 결정 -> 결정된 확률 변수로부터 다음에 생성될 데이터나 특성 예측
    - 데이터 값에서 확률변수의 확률분포함수를 역설계 하는 과정이 가장 중요
        - 데이터 분포가 가지는 표본평균, 표본 분산 등의 기술 통계값을 구한다. 
        - 이 값과 같은 기술통계값을 가지는 확률분포함수를 찾는다.
    - 데이터가 없는 확률분포 함수의 기술 통계값을 알아야 한다.
